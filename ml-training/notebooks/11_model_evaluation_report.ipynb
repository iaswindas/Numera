# 11 — Model Evaluation Report

Generates a comprehensive evaluation report for all trained models.


```python
import sys
sys.path.insert(0, "/content/drive/MyDrive/numera-ml")

from pathlib import Path
import json
import numpy as np

from google.colab import drive
drive.mount("/content/drive")

from scripts.evaluation_utils import (
    compute_classification_metrics,
    compute_matching_metrics,
    save_evaluation_report,
)

DATA_DIR = Path("/content/drive/MyDrive/numera-ml/data")
MODELS_DIR = Path("/content/drive/MyDrive/numera-ml/models")
EXPERIMENTS_DIR = Path("/content/drive/MyDrive/numera-ml/experiments")
```

```python
# Load zone classifier test predictions
zone_test_path = DATA_DIR / "annotations" / "zone_test_predictions.json"
zone_test = json.loads(zone_test_path.read_text())
true_labels = [item["true_label"] for item in zone_test]
pred_labels = [item["predicted_label"] for item in zone_test]

zone_metrics = compute_classification_metrics(true_labels, pred_labels)
print(f"Zone Classifier Accuracy: {zone_metrics['accuracy']:.4f}")
print(f"Target: >= 0.92")
```

```python
# Save combined report
combined = {
    "zone_classifier": zone_metrics,
}
save_evaluation_report(
    combined,
    str(EXPERIMENTS_DIR / "phase0_evaluation_report.json"),
    model_name="Phase 0 Demo MVP Models",
    notes="Evaluation on held-out test set",
)
print("Report saved!")
```
